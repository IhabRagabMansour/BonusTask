# Imports

```python
# Imports help the heros
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Loading Data

```python
# Load the data of our heros
Country_HQ = pd.read_excel('./Option1_Excel/Avengers_vs_Invaders_Challenge.xlsx', sheet_name='Country_HQ')
Country_HQ.head()
```

```python
# Load contacts of our heros

# we have 16 area, 12 invaders + 1 for header and 1 space after each area except the last one
# 16 * 13 + 16 - 1 = 223
contacts = pd.read_excel('./Option1_Excel/Avengers_vs_Invaders_Challenge.xlsx', sheet_name='Contacts',header=None)
contacts_dict = {}
rows_number = 14

for start_row in range(0, contacts.shape[0], rows_number):
    chunk = contacts.iloc[start_row:start_row + rows_number - 1]
    key = chunk.iloc[0, 0]
    header = chunk.iloc[0]
    chunk_df = pd.DataFrame(chunk.values[1:], columns=header)
    contacts_dict[key] = chunk_df

print(contacts_dict.keys())
invaders_species = contacts_dict['DE-Headquarter'].iloc[:,0].values
print(invaders_species)
contacts_dict['DE-Headquarter']
```

# Task 1

```python
roles = ['attack_role', 'defense_role', 'healing_role']

list_of_invaders = []

# Iterate over each row in Country_HQ DataFrame
for _, country_hq in Country_HQ.iterrows():
    # Extract country code from the second column of the current row
    country_code = country_hq.iloc[1]

    # Iterate over each invader type in the current row, starting from the third column
    for invader_type in country_hq[2:].index:
        quarter = country_hq[2:][invader_type]  # Get the quarter from the current invader_type
        quarter_df = contacts_dict[quarter]  # Access corresponding DataFrame from contacts_dict

        # Determine invader species based on invader_type
        if invader_type == 'Aliens':
            invader_species = 'aliens'
            row_index = 0
        elif invader_type == 'Predators':
            invader_species = 'predators'
            row_index = 1
        else:
            # For other types, retrieve invader names from the DataFrame and iterate through them
            invader_names = quarter_df.loc[2:][quarter].values
            for invader_species in invader_names:
                # Iterate through roles to collect emails
                for role in roles:
                    email = quarter_df[quarter_df[quarter] == invader_species][role].values[0]
                    if isinstance(email, str):
                        if '@' not in email:
                            email += '@avengers.com'
                        # Append invader details to list_of_invaders
                        list_of_invaders.append({
                            'Country_Code': country_code,
                            'Invader_Species': invader_species,
                            'Role': role,
                            'Email': email,
                            'Quarter': quarter
                        })
            continue

        # Process emails for Aliens and Predators
        for role in roles:
            email = quarter_df.iloc[row_index][role]
            if isinstance(email, str):
                if '@' not in email:
                    email += '@avengers.com'
                # Append invader details to list_of_invaders
                list_of_invaders.append({
                    'Country_Code': country_code,
                    'Invader_Species': invader_species,
                    'Role': role,
                    'Email': email,
                    'Quarter': quarter
                })

Helper_Dataframe = pd.DataFrame(list_of_invaders)
Task1 = Helper_Dataframe[['Country_Code', 'Invader_Species', 'Role', 'Email']]
Task1
```

```python
# Export File
file_path = './Option1_Excel/Avengers_vs_Invaders_Challenge.xlsx'
if os.path.exists(file_path):

    with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
        Task1.to_excel(writer, sheet_name='Task1', index=False)
else:
    with pd.ExcelWriter(file_path, engine='openpyxl') as writer:
        Task1.to_excel(writer, sheet_name='Task1', index=False)
```

# Task 2

```python
# We don't need Country_code and then we remove the duplicates
Helper_Dataframe_cleaned = Helper_Dataframe.drop(columns=['Country_Code']).drop_duplicates()
Helper_Dataframe_cleaned
```

```python
def create_role_dataframe(hero):
    # Create the new_df DataFrame
    new_df = pd.DataFrame(index=list(hero['Quarter'].unique()), columns=list(hero['Invader_Species'].unique()))
    new_df = new_df.fillna('')

    # Update new_df based on roles in DataFrame hero
    for _, row in hero.iterrows():
        invader_species = row['Invader_Species']
        quarter = row['Quarter']
        role = row['Role']
        role_char = 'A' if role == 'attack_role' else 'D' if role == 'defense_role' else 'H'
        new_df.loc[quarter, invader_species] += role_char

    return new_df

def process_emails(df):
    unique_emails = df['Email'].unique()
    heros_dict = {}

    for email in unique_emails:
        email_data = df[df['Email'] == email]
        new_df = create_role_dataframe(email_data)

        name = email.split('@')[0]
        new_df.index.name = name
        new_df.reset_index(inplace=True)
        heros_dict[name] = new_df

    return heros_dict
```

```python
Task2 = process_emails(Helper_Dataframe_cleaned)
```
